# XYZ Company has offices in four different zones. The company wishes to investigate the following :                                                                                           ● The mean sales generated by each zone.                                                                ● Total sales generated by all the zones for each month.                                          ● Check whether all the zones generate the same amount of sales.                  Help the company to carry out their study with the help of data provided.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('Sales_data_zone_wise.csv')
df
```

# ● The mean sales generated by each zone.  

```python
zam=df['Zone - A'].mean()
zbm=df['Zone - B'].mean()
zcm=df['Zone - C'].mean()
zdm=df['Zone - D'].mean()
print("The mean sales generated by zone A = ",zam)
print("The mean sales generated by zone B = ",zbm)
print("The mean sales generated by zone C = ",zcm)
print("The mean sales generated by zone D = ",zdm)
```

# Total sales generated by all the zones for each month.  

```python
df['Total monthly sale']= df['Zone - A']+df['Zone - B']+df['Zone - C']+df['Zone - D']
df
```

```python
plt.figure(figsize=(20,10))
x=df['Month']
y=df['Total monthly sale']
sns.barplot(data =df ,x="Month", y="Total monthly sale")
plt.title("Total Sales by Month", fontsize = 15)
plt.xlabel("Months", fontsize=12)
plt.ylabel("Total Monthly Sale", fontsize = 10)
plt.xticks(rotation = 90)
for i in range(len(x)):
    plt.text(i,y[i],y[i],ha="center",va="bottom")
```

# Check whether all the zones generate the same amount of sales.                  

H0: All zones generate same amount of sales
H1: All zones generate different amount of sales

```python
from scipy.stats import f_oneway

fvalue, pvalue = f_oneway(df['Zone - A'], df['Zone - B'], df['Zone - C'],df['Zone - D'])
print("The F-statistic value is : " ,fvalue)
print("The Calculated p value is : " ,pvalue)

alpha = 0.05
if pvalue < alpha:
    print('Rejecting H0')
else:
    print ('Failing to reject H0')
```

Based on the test carried out on the provided data we are failing to accept the null hypoyhesis.
"All zone generate different amount of sales"
